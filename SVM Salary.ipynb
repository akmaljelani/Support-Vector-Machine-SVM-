```python
import pandas as pd 
import numpy as np 
import seaborn as sns
```

```python
train=pd.read_csv('SalaryData_Train (3).csv')
test=pd.read_csv('SalaryData_Test (3).csv')
```

```python
train.head()
```

```python
test.head()
```

```python
string_columns = ["workclass","education","maritalstatus","occupation","relationship","race","sex","native"]
```

```python
##Preprocessing the data. As, there are categorical variables
from sklearn.preprocessing import LabelEncoder
number = LabelEncoder()
for i in string_columns:
        train[i]= number.fit_transform(train[i])
        test[i]=number.fit_transform(test[i])
```

```python
test
```

```python
train
```

Now all string values(categorical) have been converted

```python
## Understanding columns
colnames = train.columns
colnames
```

```python
len(colnames)
```

```python
## defining X & Y variables for the columns in training & test data

x_train = train[colnames[0:13]]
y_train = train[colnames[13]]
x_test = test[colnames[0:13]]
y_test = test[colnames[13]]
```

```python
## Normalizing function to normalize all the numerical values that have been converted
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
x_train = norm_func(x_train)
x_test =  norm_func(x_test)
```

```python
x_train
```

```python
x_test
```

Values for X have been normalized

```python
## Importing Support Vector Machine
from sklearn.svm import SVC
```

```python
## using linear kernel SVM

model_linear = SVC(kernel = "linear")
model_linear.fit(x_train,y_train)
pred_test_linear = model_linear.predict(x_test)
```

```python
np.mean(pred_test_linear==y_test)
```

Accuracy for test data using linear Kernel SVM is 80.9% accurate

```python
## using rbf kernel SVM

model_rbf = SVC(kernel = "rbf")
model_rbf.fit(x_train,y_train)
pred_test_rbf = model_rbf.predict(x_test)
```

```python
np.mean(pred_test_rbf==y_test)
```

Accuracy for test data using rbf Kernel SVM is 84.32% accurate

```python
## using poly kernel SVM

model_poly = SVC(kernel = "poly")
model_poly.fit(x_train,y_train)
pred_test_poly = model_poly.predict(x_test)
```

```python
np.mean(pred_test_poly==y_test)
```

Accuracy for test data using poly Kernel SVM is 84.35% accurate

```python
## using sigmoid kernel SVM

model_sig = SVC(kernel = "sigmoid")
model_sig.fit(x_train,y_train)
pred_test_sig = model_rbf.predict(x_test)
```

```python
np.mean(pred_test_sig==y_test)
```
