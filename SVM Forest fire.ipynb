```python
import pandas as pd 
import numpy as np 
import seaborn as sns
```

```python
forestfire=pd.read_csv('forestfires (3).csv')
forestfire.head()
```

```python
forestfire.describe()
```

```python
forestfire.info()
```

```python
## Dropping month & day from dataframe
forestfire.drop(["month","day"],axis=1,inplace =True)
```

```python
forestfire.head()
```

```python
## Locating predictors & targets
predictors = forestfire.iloc[:,0:28]
target = forestfire.iloc[:,28]
```

```python
## normalizing function
def norm_func(i):
    x= (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
forestfire_n = norm_func(predictors)
```

```python
forestfire_n.head()
```

```python
## importing Support Vector Machine

from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(predictors,target,test_size = 0.20, stratify = target)
```

```python
x_train
```

```python
x_test
```

```python
## using linear kernel SVM

model_linear = SVC(kernel = "linear")
model_linear.fit(x_train,y_train)
pred_test_linear = model_linear.predict(x_test)
```

```python
np.mean(pred_test_linear==y_test) 
```

Accuracy for training data using linear Kernel SVM is 99% accurate

```python
## using linear kernel SVM

model_poly = SVC(kernel = "poly")
model_poly.fit(x_train,y_train)
pred_test_poly = model_poly.predict(x_test)
```

```python
np.mean(pred_test_poly==y_test)
```

Accuracy for test data using poly Kernel SVM is 75% accurate

```python
## using rbf kernel SVM

model_rbf = SVC(kernel = "rbf")
model_rbf.fit(x_train,y_train)
pred_test_rbf = model_rbf.predict(x_test)
```

```python
np.mean(pred_test_rbf==y_test)
```

Accuracy for test data using rbf Kernel SVM is 74% accurate

# For this data, using linear kernel SVM has a 99% accuracy in predicting forest fires.
